```python
def winner(points):
    if points > 580:
        print("First place")
    elif points <= 526:
        print("Second place")
    elif points == 362:
        print("Third place")
    else:
        print("Not in the top three")
    return(points)
```

```python
points_spain = 583
winner(points_spain)
```
