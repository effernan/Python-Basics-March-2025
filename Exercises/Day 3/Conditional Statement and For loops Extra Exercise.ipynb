As you can observe, the condition of winning a match is to have a result higher than the number of goals scored by your team's rival. Write a function that abstracts that concept and test it introducing match results of the 2022 world cup.

```python
#Your code in here
```

```python
def winner(x, y):
    if x > y:
        print(f"{x} is the winner")
    elif x == y:
        print("match")
    else:
        print(f"{x} is the looser")
    return(winner)
```

Now that you have all that data under your belt, FIFA has given you a new task: you are supposed to predict how many goals you have to score to be a winner in each phase (round 16, quarters, semi-final, final) in the next 2026 world cup using the 2018 and the 2022 world cup data. <br>

To do this, you are going to use a simple method used in Statistics called "Linear Regression". A linear regression uses two variables (2018 winners, 2022 winners) to predict a third one: 2026 winners. 
 
The data of the 2018 world cup (https://en.wikipedia.org/wiki/2018_FIFA_World_Cup#Knockout_stage) has been conveniently prepared for you. You can find the two lists (2018, 2022) in here:

```python
winners_scores_round_16_2022 = [3, 2, 3, 4, 3, 3, 3, 6]
winner_scores_quarter_finals_2022 = [3, 4, 4, 2]
winner_scores_semi_finals_2022 = [3, 2]
winner_scores_final_2022 = [4]
```

```python
winners_scores_round_16_2022
```

```python
winners_scores_round_16_2018 = [2, 4, 2, 3, 3, 3, 1, 4]
winner_scores_quarter_finals_2018 = [2, 2, 4, 2]
winner_scores_semi_finals_2018 = [2, 1]
winner_scores_final_2018 = [4]
```

Let's practice using the round 16 data from 2022 and 2018. After that, repeat the exercise with quarter finals, semi finals, and final.

```python
# first we import the libraries

import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# then we feed the data to the linear regression 

x = np.array(winners_scores_round_16_2022).reshape((-1, 1))
y = np.array(winners_scores_round_16_2018)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
intercept: 5.633333333333329

print(f"slope: {model.coef_}")
slope: [0.54]
```

```python
new_model = LinearRegression().fit(x, y.reshape((-1, 1)))

print(f"intercept: {new_model.intercept_}")
intercept: [5.63333333]

print(f"slope: {new_model.coef_}")
slope: [[0.54]]
```

```python
#and now we have the predicted response!

y_pred = model.predict(x)

print(f"predicted response:\n{y_pred}")
```
