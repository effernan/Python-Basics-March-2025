## 2.2. Bar Plots

Different type of visualizations offer different options depending of our data needs. If we would like to represent our Barbie dataset using a bar plot, we would need to change our focus of analysis. 

First of all: let's ask our language model how to do a bar pot (https://www.perplexity.ai/search/d1f4b22e-dd17-483b-b199-6539d3f00dee?s=u).

So, basically: we need two variables (X and Y). With our dataframe, we have two options.

**a) We explore things at the Career level.** For example: let's have a look at the STEM dataframe.

```python
#Your code here
```

```python
#Solution
STEM
```

```python
plt.bar(STEM["Year"], STEM["Career"], color = "blue")
plt.title('Barbie STEM')
plt.xlabel('Year')
plt.ylabel('Career')
plt.show()
```

That works, but is not particularly useful. We could use plot bars though to see quite fast which Type of career is the most popular within barbie. To do that, we first need to create a new dataset where we have career unique counts. Let's ask our language model how to do that: https://www.perplexity.ai/search/6458f8d8-94e9-4ea6-8035-b86032cd0478?s=u

```python
#Your code here
```

```python
#Solution
dups = barbie.pivot_table(index=['Type'], aggfunc='size')
```

```python
dups
```

So: let's create a new dataframe with those values!

```python
#Your code in here
```

```python
#Solution

career = ["Arts", "Beauty", "Business", "Education", "Medicine", "Military", "Political", "Public Service", 
         "STEM", "Sport", "Transportation"]
counts = [22, 67, 25, 23, 11, 8, 9, 7, 12, 31, 15]
```

```python
df = pd.DataFrame(list(zip(career, counts)), columns=['Career', 'Counts'])

df
```

And now let's do the Bar plot!

```python
# Your code in here
```

```python
#Solution


plt.figure(figsize = (20, 10)) #this is to make the plot the size that we want

plt.bar(df["Career"], df["Counts"], color = "pink")
plt.title('Barbie Careers')
plt.xlabel('Year')
plt.ylabel('Career')

plt.show()
```

So: it is clear that the type of profession that is most popular among barbie dolls is the Beauty industry! It is also clear that we need ot choose our data visualization depending on our dataset and on our research question. We can also combine different types of data visualizations to have a better understanding of data patterns!
